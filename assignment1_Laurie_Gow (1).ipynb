d. Print for each of the numbers x = 2; : : : 20, all numbers that divide x, excluding 1 and x. Hence,
for 18, it should print 2 3 6 9.

```python
#set up a list to print the divisors for each number
def print_divisors(x):
    divisors = [i for i in range(2,x) if x % i == 0]
    print(f"The divisors of {x} are:", *divisors)

#loop to call the divisors
for x in range(2, 21):
    print_divisors(x)
```

e. The smallest number that is divisible by 2, 3 and 4 is 12. Find the smallest number that is divisible
by all integers between 1 and 10.


```python
# import necessary library
import math

# Returns the lcm of first 10 numbers
def lcm(n):
    ans = 1
    for i in range(1, n + 1):
        ans = int((ans * i)/math.gcd(ans, i))        
    return ans

# print the answer
n = 10
print ("The smallest number divisible by all integers between 1 and 10 is:",lcm(n))
```

a. Calculate A + A, AA^T, A^TA, and AB.

```python
# install necessary packages
import numpy as np
from scipy.linalg import toeplitz
import time

# generate matrices A and B and print
A=np.round(np.random.normal(size=(200,500)))
print('A=')
print(A)

B=toeplitz(range(1,501))
print('\nB=')
print(B)

# calculate and print A+A
print('\nA+A=')
print(A+A)

# generate A transpose and print
A_Transpose=np.transpose(A)
print('\nA_Transpose=')
print(A_Transpose)

# calculate and print AA^T
print('\nAA^T=')
print(A.dot(A_Transpose))

# calculate and print A^TA
print('\nA^TA=')
print(A_Transpose.dot(A))

# caluculate AB and print
print('\nAB=')
print(A.dot(B))
```

c. Nearest neighbor. Write a for loop that takes a value z and an array A and finds the element in A
that is closest to z. it should print the closest value, not index.
Hint: Use the built-in functionality of Numpy rather than writing code to find this value manually.
In particular, use brackets and argmin.


```python
def nearestneighbors(A,Z):
    c=np.abs(A-Z)
    index=np.argmin(c)
    return A[index]
A=np.array(range(1,100))

nearestneighbors(A,5)   
```
